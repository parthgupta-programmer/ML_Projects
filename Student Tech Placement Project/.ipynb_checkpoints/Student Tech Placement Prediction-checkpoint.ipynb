## Importing required libraries and data

```python
# Importing all libraries which is required 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Importing the data from a csv file

dataset=pd.read_csv(r"C:\Users\HP\Desktop\PG\ML_Projects\Student Tech Placement Project\student_data_raw_2000.csv")
```

## Handling Data

```python
# Checking the no.of null values contained in each particular feature

dataset.isnull().sum()
```

```python
# Deleting the rows where the Placed Tech is filled with N.A.

dataset.dropna(subset=['Placed_Tech'],inplace=True)
```

```python
# Dataset after deleting N.A. of placed tech.

dataset
```

```python
print(dataset['Placed_Tech'].value_counts())
```

```python
# Checking and correcting the City names 

dataset['City'].unique()
```

```python
# Replace function that standardizes city names  (handles different spellings, cases, and extra spaces)

def replace(string):
    if(string=='NEW DELHI' or string=='New  Delhi'):
        return 'New Delhi'
    elif(string=='ghaziabad' or string=='Gaziabad'):
        return 'Ghaziabad'
    elif(string=='gorakhpur'):
        return 'Gorakhpur'
    elif(string=='Roorkee '):
        return 'Roorkee'
    else:
        return string
dataset['City']=(dataset['City'].apply(replace))
```

```python
# Dataset after Standardizing city names

dataset
```

```python
# Now , the city names became unique.

dataset['City'].unique()
```

```python
# Converting any negative values in Deep_Learning_Hours to absolute values.

dataset['Deep_Learning_Hours']=(dataset['Deep_Learning_Hours'].apply(lambda n: abs(n)))
dataset
```

```python
# Filling null values of C++ Score,Deep_Learning_Hours and Cloud_Hackathon

dataset['C++_Score'].fillna(dataset['C++_Score'].median(),inplace=True)
dataset['Deep_Learning_Hours'].fillna(dataset['Deep_Learning_Hours'].median(),inplace=True)
dataset['Cloud_Hackathon'] = dataset.groupby('Placed_Tech')['Cloud_Hackathon'].transform(lambda x: x.fillna(x.mode()[0]))


dataset['Cloud_Hackathon'] = dataset['Cloud_Hackathon'].str.lower().map({
    'yes': 1, 'y': 1,
    'no': 0, 'n': 0
})

dataset['Cloud_Hackathon'].unique()
```

```python
# Dataset after the filling null values

dataset
```

## Identifying and removing outliers

```python
# Description of the data

dataset.describe()
```

```python
# Plotting the scatter chart between the Deep Learning hours and C++ score and Placed_Tech as a Hue parameter.

%matplotlib inline
plt.figure(figsize=(12,8))
sns.scatterplot(x='C++_Score',y='Deep_Learning_Hours',data=dataset,hue='Placed_Tech');
```

```python
# Finding outlier for C++_Score column

Q1=dataset['C++_Score'].quantile(0.25)
Q3=dataset['C++_Score'].quantile(0.75)

IQR=Q3-Q1

lower_fence = Q1-(1.5*IQR)
upper_fence = Q3+(1.5*IQR)

print("Upper Fence: ",upper_fence)
print("Lower Fence: ",lower_fence)
```

```python
# Cleaned dataset 

cleaned_dataset=dataset[(dataset['C++_Score'] > lower_fence) &  (dataset['C++_Score'] < upper_fence)]
cleaned_dataset # Dataset after the removal of the outliers
```

## Data Preprocessing

```python
# Seperating features in X and target in y

X=cleaned_dataset[['C++_Score','Deep_Learning_Hours','Cloud_Hackathon']]
y=cleaned_dataset['Placed_Tech']
```

```python
# It is always recommended first to split the data into training data and testing data before inspecting on that data.

# Here, test ratio=0.25 means that 25% of data is seperated to test the data/model and rest 75% is for training the data/model.

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
# print(X_train)
```
